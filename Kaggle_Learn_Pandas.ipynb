# Kaggle Learn: Pandas コース メソッド・関数一覧

---

## 1. Creating, Reading and Writing（データの作成・読み込み・書き込み）
データ分析の起点となる、データの生成と外部ファイルの入出力に関する機能です。

* **`pd.DataFrame()`**：辞書データなどから2次元のデータフレーム（`DataFrame`）を作成する。
* **`pd.Series()`**：リストなどから1次元のデータ配列（`Series`）を作成する。
* **`pd.read_csv()`**：CSVファイルを読み込んでデータフレームに変換する。
* **`df.shape`**：データフレームの行数と列数をタプル `(行数, 列数)` で返す属性。
* **`df.head()`**：データの先頭5行（引数で指定した行数）を返す。
* **`df.to_csv()`**：データフレームをCSVファイルとして保存する。


---

## 2. Indexing, Selecting & Assigning（インデックス、選択、代入）
データフレームから特定の行や列、ピンポイントの要素を抽出・変更する機能です。

* **列の選択（`df.column_name` / `df['column_name']`）**：特定の列をSeriesとして抽出する。
* **`df.iloc[]`**：**行列の番号（整数インデックス値）**を指定してデータを抽出する（`iloc[行番号, 列番号]`）。
* **`df.loc[]`**：**ラベル名（インデックス名やカラム名）**を指定してデータを抽出する（`loc[行名, 列名]`）。

　　　loc の構造は [ WHERE , SELECT ]

　　　loc の角括弧の中身は、カンマを挟んで左側が行（WHERE）、右側が列（SELECT）の役割を持っています。

　　　行とカラム指定して抜き出したいときはこっち。それ以外はリスト形式

　　　値を置き換えたいときも直接指定できるこっち

* **`df.set_index()`**：指定した列をデータフレームのインデックス（行名）に設定する。


---

## 3. Summary Functions and Maps（要約関数と写像）
データの全体像を把握するための統計量の算出や、データの一括変換を行う機能です。

* **`df.describe()`**：数値列の平均、標準偏差、最大値、最小値などの要約統計量をまとめて出力する（文字列データの場合は最頻値や種類数を出力）。
* **`s.mean()`**：Seriesの平均値を算出する。
* **`s.unique()`**：Seriesに含まれる重複のない固有の値を返す。
* **`s.value_counts()`**：値ごとの出現回数をカウントし、多い順に返す。
* **`s.map()`**：Seriesの全要素に対して、特定の関数や辞書を適用して値を変換（写像）する。
* **`df.apply()`**：データフレームの行方向または列方向（`axis`で指定）に対して一括して関数を適用する。



---

## 4. Grouping and Sorting（グルーピングとソート）
特定のカテゴリごとにデータをグループ化して集計したり、特定の規則で並び替えたりする機能です。

* **`df.groupby()`**：指定した列の値ごとにデータをグループ化する（集計関数と組み合わせて使用）。
* **`s.groupby().agg()` / `df.groupby().agg()`**：複数の異なる集計処理（`min`, `max`, `count` など）を一度に適用する。
* **`df.sort_values()`**：指定した列の値を基準にデータを並び替える（昇順・降順の設定が可能）。
* **`df.sort_index()`**：インデックス（行名）を基準にデータを並び替える。



---

## 5. Data Types and Missing Values（データ型と欠損値）
各列のデータ型の確認・変更や、データ内に含まれる欠損値（NaN）の検出・穴埋めを行う機能です。

* **`df.dtypes` / `s.dtype`**：データフレーム各列、またはSeriesのデータ型（`int64`, `float64`, `object` など）を確認する属性。
* **`s.astype()`**：データ型を明示的に変換する（例：整数型から浮動小数点型への変換）。
* **`df.isnull()` / `df.isna()`**：欠損値（NaN）であるかどうかを判定し、ブール値（True/False）のマスクを返す。
* **`df.fillna()`**：欠損値（NaN）を指定した値（0や平均値など）で穴埋めする。
* **`df.replace()`**：特定の値を別の値に置換する。



---
6. Renaming and Combining（名称変更と結合）

名称変更（Renaming）
データフレームの列名（カラム名）や行名（インデックス名）を任意の名前に変更する機能です。

* **`df.rename()`**：辞書型（{'旧名': '新名'}）を用いて、特定の列名や行名を個別に変更する。

* **`df.rename_axis()`**：文字列などを指定して、行や列の軸全体の見出し（タイトル名）を変更する。

* **`df.set_axis()`**：新しい名称のリストを渡すことで、軸（列または行）全体の名前を一度に上書き変更する。

結合（Combining）
複数の異なるデータフレームを、共通のキーやインデックスをベースにつなぎ合わせる機能です。
* **`pd.concat()`**：複数のデータフレームを軸方向（縦方向または横方向）に単純に連結する。
* **`df.join()`**：インデックス（行名）を基準にして、別のデータフレームを左結合・右結合などで結合する。
* **`df.merge()`**：SQLのJOIN操作のように、共通する特定の「列（キー）」を基準にしてデータフレームを結合する。

````python
# Kaggleコースにない「日付・タイムスタンプ（Datetime）」最重要メソッド

Pandasで日付データを扱うための、実務・コンペ必須の基本操作です。
Colabの住宅データには日付がないため、ここではコードを実行するとその場でダミーの日付データを作って試せるようにしています。

---

## 1. pd.to_datetime() —— 文字列を「日付型」に変換する
CSVなどを読み込んだ直後、日付の列はただの「文字列（object型）」になっています。これをPandasが日付として認識できる「datetime64型」に変換する超最重要メソッドです。

```python
import pandas as pd

# 実験用のダミーデータ（文字列の日付）
df_time = pd.DataFrame({
    'date_str': ['2026-05-01 09:00:00', '2026-05-02 18:30:00', '2026-05-03 23:15:00']
})
print("変換前（型を確認）:")
print(df_time['date_str'].dtype) # 'object' (文字列) と表示される

# 日付型に変換
df_time['date_dt'] = pd.to_datetime(df_time['date_str'])

print("\n変換後（型を確認）:")
print(df_time['date_dt'].dtype) # 'datetime64[ns]' と表示される
````

.dt アクセサ —— 年・月・日・曜日・時間をぶっこ抜く
列が「日付型（datetime64）」になっていれば、.dt を後ろにつけるだけで、年や月、曜日などを一瞬で新しい列として抽出できます。

s.dt.year ： 年

s.dt.month ： 月

s.dt.day ： 日

s.dt.hour ： 時間

s.dt.day_name() ： 曜日名（Monday, Tuesdayなど）

````python
# 日付型から「月」「時間」「曜日」を抽出して新しい列を作る
df_time['month'] = df_time['date_dt'].dt.month
df_time['hour'] = df_time['date_dt'].dt.hour
df_time['day_of_week'] = df_time['date_dt'].dt.day_name()

display(df_time)
````

日付の引き算（TimeDelta） —— 期間（あと何日？）を計算する
日付型のデータ同士は、そのまま引き算ができます。計算結果は TimeDelta という「期間」を表す特殊な型になります。

````python
# 例：ターゲットとなる締め切り日
deadline = pd.to_datetime('2026-05-31 00:00:00')

# 締め切りまで「あと何日と何時間あるか」を計算
df_time['time_to_deadline'] = deadline - df_time['date_dt']

# .dt.days を使うと「日数」だけを整数で取り出せる
df_time['days_remaining'] = df_time['time_to_deadline'].dt.days

display(df_time[['date_dt', 'time_to_deadline', 'days_remaining']])
````
