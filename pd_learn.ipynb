```python
import pandas as pd

# このデータフレーム（sensor_df）を使います！
sensor_df = pd.DataFrame({
    'temp': [22.5, 45.0, 18.2, 50.5],
    'press': [100, 150, 95, 210]
})
# 1. 判定用の関数を作る
def check_sensor(row):
    if row['temp']>=50 or row['press']>=200:
        return 1
    else:
        return 0

# 2. apply で適用する
sensor_df['error_flag'] = sensor_df.apply([check_sensor],axis=1)  #axis=0 縦方向に切り出し、インデックスの数字しか見えないため'temp'や'press'指定はエラー
sensor_df                                                         #axis=1　横方向に切り出しシリーズ化、逆にインデックスの数字は見えないため、カラム名指定

#Pandasの Series（1本の列）というデータ構造は、「値」と「インデックス」の2つしか持てない というケチな制約があります。
#そのため、切り出した瞬間に、どちらか片方の情報（行名か列名か）をインデックスとして採用し、もう片方は捨てざるを得ない、という大人の事情（仕様）からこの動きになっています。
```

```python
import pandas as pd

d={'items_name':['りんご', 'バナナ', 'オレンジ', 'メロン'],'price':[150, 100, 120, 800],'sold_count':[10, 25, 15, 3]}
df=pd.DataFrame(d)
translation_dict = {
    'りんご': 'Apple',
    'バナナ': 'Banana',
    'オレンジ': 'Orange',
    'メロン': 'Melon'
}
df['english_name']=df['items_name'].map(translation_dict)  #カラム一つに適用
mpr=df['price'].mean()
df['npr']=df['price'].map(lambda p: p-mpr)
df2=df[['price','sold_count']].apply(sum)  #すべてのカラムにSUM適用
df2
```
